```python
x = np.array([2.6, 3.0, 3.2, 3.6, 4.0])
y = np.array([55, 56.5, 61, 68, 72.5])
plt.scatter(x, y)
plt.show()

model = LinearRegression()
model.fit(x.reshape(-1, 1), y.reshape(-1, 1))
print(model.coef_)
print(model.intercept_)
y = 18.06 + 13.58 * x

y_prediction1=model.coef_*x+model.intercept_
plt.scatter(x, y)
plt.plot(x,y_prediction1.reshape(-1, 1),color='r')
plt.show()

y_prediction2=(model.coef_-3)*x+model.intercept_+10
plt.scatter(x, y)
plt.plot(x,y_prediction2.reshape(-1, 1),color='y')
plt.show()

y_prediction3=(model.coef_-8)*x+model.intercept_+25
plt.scatter(x, y)
plt.plot(x,y_prediction3.reshape(-1, 1),color='g')
```

* The best fit line is the one for which total prediction error (all data points) are as small as possible.
* The error is the mean of square errors
* (MSE or mean of SSE) between the point to the predictions on the regression line.

### Gradient descent
* Gradient descent is an algorithm that finds best fit line for given training data points.
* How to get the best Y = 18.06 + 13.58 * X?
* How to get the minimum errors?
    * min(MSE)or min(1/n∑ (y - ŷ )^2)
* Gradient descent is an algorithm that finds best fit line for given training data points

* MSE = 1/n∑ (y-(β+α*x)))^2
* ∂/∂β = -2/n∑ x(y-(β+α*x))
* ∂/∂α = -2/n∑ (y-(β+α*x))

* β=β-r*∂/∂β
* α=α-r*∂/∂α

```python
def gradient_descent(x, y):
    slope=intercept = 0
    iterations = 5000
    n = len(x)
    rate = 0.05
    for i in range(iterations):
        y_predict = slope*x + intercept
        mse = (1/n)*(y-y_predict).dot(y-y_predict)
        d_slope = -(2/n)*(x.dot(y-y_predict))
        d_intercept = -(2/n)*sum(y- y_predict)
        slope -= rate*d_slope
        intercept -= rate*d_intercept
        print("iteration= {}, slope= {}, intercept= {}, mse= {}".format(i, slope, intercept, mse))
gradient_descent(x, y)
```
