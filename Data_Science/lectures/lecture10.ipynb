### standard normal distribution
* ![](https://s3.studylib.net/store/data/009424308_1-ca24100c56a2ceddad5f64dc65fde1ce.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x=np.linspace(-5,5,10**3)
plt.plot(x, norm.cdf(x))
plt.xlabel('x')
plt.ylabel('cumulative probability')
plt.title('Standard Normal Distribution CDF')
plt.show()
```

```python
print(norm.cdf(0.5)-norm.cdf(-0.5))
print(2*(0.5-norm.cdf(-0.5)))
print(2*(norm.cdf(0)-norm.cdf(-0.5)))
```

```python
#95% confidence interval
print(norm.cdf(1.96) - norm.cdf(-1.96))
print(1-2*norm.cdf(-1.96))
```

### normal distribution
* The US National Center for Chronic Disease Prevention and Health Promotion
surveyed >400,000 individuals for health-related info (BRFSS – Behavioral Risk
Factor Surveillance System)
* The distribution is roughly normal with parameter µ = 178cm and σ^2=59.4cm^2
    * σ=sqrt(59.4) = 7.707cm


```python
Image(filename=r'C:\Users\rmani\OneDrive\Pictures\htmls.png')
```

* What percentage of US male population is between 5’10” and 6’1”?
* 5’10” = 177.8cm; 6’1” = 185.4cm
* Height distribution is in X ~ N(178, 59.4)
* P(177.8 £ X £ 185.4) = ?

```python
print(norm.cdf(185.4,178,7.707)-norm.cdf(177.8, 178, 7.707))

print(norm.cdf(7,5,3)-norm.cdf(2, 5, 3))
```

* X’ = (X – 178) / 7.707 ~ N(0, 1)
    * (185.4 – 178) / 7.707 = 0.96
    * (177.8-178)/7.707 = -0.03

```python
print(norm.cdf(0.96) - norm.cdf(-0.03))
```

### Why model
* Data compression – a small set of parameters may be sufficient to
summarize a large data set
* Sometimes can smooth out noises
* When data from a natural phenomenon fit a distribution, it can lead to insight into the physical system which can explain why the observed data has a particular form
* **Many machine learning methods assume certain data distribution. It is important to investigate the actual data distribution**
    * It is often difficult to fit an exact model.
    * Make reasonable approximations
    * Try different transformation and cleaning
